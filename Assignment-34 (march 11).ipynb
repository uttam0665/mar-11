```python
Q2: Differentiate between one-tailed and two-tailed tests.
 
 ANS:=
        
1.One-Tailed Test:

In a one-tailed test, the hypothesis being tested is directional, meaning it focuses on whether the observed data is significantly greater than or less than a certain value.
The critical region is located in only one tail of the distribution.
It is used when there is a specific expectation or hypothesis about the direction of the effect or relationship being tested.
The p-value obtained from a one-tailed test represents the probability of observing the data in the specified direction under the null hypothesis.
Example: A researcher wants to test whether a new drug increases the average response time of participants. The researcher expects that the drug will lead to slower response times. In this case, a one-tailed test is appropriate as the hypothesis is directional, focusing on the drug's effect in one direction (slower response times).

2.Two-Tailed Test:

In a two-tailed test, the hypothesis being tested is non-directional, meaning it examines whether the observed data is significantly different from a certain value, without specifying the direction.
The critical region is divided into two equal tails of the distribution.
It is used when there is no specific expectation or hypothesis about the direction of the effect or relationship being tested.
The p-value obtained from a two-tailed test represents the probability of observing the data as extreme or more extreme in both directions (greater than or less than) under the null hypothesis.
Example: A researcher wants to test whether a new teaching method improves test scores. The researcher does not have a specific expectation about whether the scores will increase or decrease. In this case, a two-tailed test is appropriate as the hypothesis is non-directional, examining whether there is any significant difference in test scores.

The choice between a one-tailed and two-tailed test depends on the research question and the specific hypotheses being tested.




     
```

```python
Q3: Explain the concept of Type 1 and Type 2 errors in hypothesis testing. Provide an example scenario for
each type of error.

Ans:=
    
1.Type 1 Error:

.Type 1 error, also known as a false positive, occurs when the null hypothesis is incorrectly rejected, indicating a significant effect or relationship when there is actually no such effect or relationship in the population.
.It is the error of concluding that there is a difference or an effect when none truly exists.
.The significance level (alpha) determines the probability of committing a Type 1 error, typically set at 0.05 or 0.01.
.Type 1 errors are more concerning in situations where making a false positive conclusion has serious consequences.

Example Scenario for Type 1 Error:
Suppose a pharmaceutical company is testing a new drug for a particular disease. The null hypothesis states that the drug has no effect. In reality, the drug is ineffective and does not improve the condition. However, in the study, due to random sampling or other factors, the data appears to show a significant improvement in patients who received the drug. If the researchers reject the null hypothesis and conclude that the drug is effective (when it's not), it would be a Type 1 error.

2.Type 2 Error:

.Type 2 error, also known as a false negative, occurs when the null hypothesis is incorrectly accepted, failing to detect a true effect or relationship that exists in the population.
.It is the error of failing to identify a difference or an effect when one truly exists.
.Type 2 errors are influenced by factors such as sample size, effect size, and the chosen level of significance.
.Type 2 errors are more concerning in situations where failing to detect a true effect can have significant consequences.

Example Scenario for Type 2 Error:
Consider a diagnostic test for a disease. The null hypothesis states that the individual does not have the disease, while the alternative hypothesis states that they do have the disease. In reality, the individual does have the disease, but the test fails to detect it, providing a negative result. If the researchers fail to reject the null hypothesis and conclude that the individual does not have the disease (when they do), it would be a Type 2 error.

It's important to note that Type 1 and Type 2 errors are inversely related, meaning reducing the probability of one type of error often increases the probability of the other type. The goal is to strike a balance based on the specific context and consequences of each error.




    
```

```python
Q4:  Explain Bayes's theorem with an example
 
  Ans:= 
        
                        
  Bayes's theorem is a fundamental concept in probability theory and statistics that allows us to update our beliefs or probabilities about an event based on new evidence. It provides a mathematical framework for incorporating prior knowledge and updating it with observed data.

Bayes's theorem can be stated as follows:

P(A|B) = (P(B|A) * P(A)) / P(B)

Where:

P(A|B) is the posterior probability of event A given event B.
P(B|A) is the conditional probability of event B given event A.
P(A) is the prior probability of event A.
P(B) is the prior probability of event B.
To understand Bayes's theorem, let's consider an example:

Example:
Suppose there is a certain medical condition that affects 1% of the population. There is a diagnostic test for this condition that has a 95% accuracy rate, meaning it correctly identifies the presence of the condition in 95% of cases where it is present (true positive rate) and correctly identifies the absence of the condition in 95% of cases where it is not present (true negative rate).

Now, imagine a person tests positive for the condition using this test. The question is, what is the probability that the person actually has the condition?

Let's apply Bayes's theorem to this scenario:

P(A) = 0.01 (prior probability of having the condition)
P(B|A) = 0.95 (probability of testing positive given the person has the condition)
P(B) = ? (probability of testing positive, which needs to be calculated)

To calculate P(B), we need to consider both true positives and false positives:
P(B) = P(B|A) * P(A) + P(B|not A) * P(not A)
= 0.95 * 0.01 + (1 - 0.95) * (1 - 0.01)
= 0.01 + 0.99 * 0.05
= 0.01 + 0.0495
= 0.0595

Now, we can calculate the posterior probability using Bayes's theorem:
P(A|B) = (P(B|A) * P(A)) / P(B)
= (0.95 * 0.01) / 0.0595
= 0.0095 / 0.0595
≈ 0.1597

So, the probability that the person actually has the condition given a positive test result is approximately 0.1597 or 15.97%.

Bayes's theorem allows us to update our initial belief (prior probability) based on the observed data (conditional probability). 
In this example, even with a relatively accurate test, the low prevalence of the condition results in a relatively low probability of having the condition even after testing positive.




      
                
```

```python
Q5: What is a confidence interval? How to calculate the confidence interval, explain with an example.
    
    Ans:=
        
    A confidence interval is a range of values that provides an estimated range of plausible values for an unknown population parameter, such as a mean or proportion. It is used to quantify the uncertainty associated with estimating the true value based on a sample.

To calculate a confidence interval, the following steps are typically followed:

1.Determine the confidence level: The confidence level reflects the level of confidence that the true population parameter falls within the calculated interval. Commonly used confidence levels are 90%, 95%, and 99%.

2.Collect a sample: Obtain a representative sample from the population of interest. The sample should be randomly selected and sufficiently large to meet the assumptions of the statistical test being used.

3.Compute the sample statistic: Calculate the sample statistic that represents the parameter of interest. For example, if estimating a population mean, the sample mean is computed.

4.Determine the margin of error: The margin of error represents the maximum amount that the estimate may differ from the true population parameter. 
It is calculated based on the standard error of the sample statistic and depends on the desired confidence level.

5.Calculate the confidence interval: The confidence interval is constructed by taking the point estimate (e.g., sample mean) and adding/subtracting the margin of error. 
The resulting range provides the lower and upper bounds of the interval.

Now, let's illustrate with an example:

Suppose you want to estimate the average height of a certain species of plants. You collect a random sample of 100 plants and measure their heights. The sample mean height is 50 cm, and the sample standard deviation is 5 cm.

1.Assume a desired confidence level of 95%.

Collect a sample of 100 plants and measure their heights.

Compute the sample mean: 50 cm.

Determine the margin of error: This depends on the sample size, standard deviation, and desired confidence level. For simplicity, let's assume a standard deviation of 5 cm and calculate the margin of error using a t-distribution (since the population standard deviation is unknown). 
    With a sample size of 100 and a 95% confidence level, the margin of error can be determined from the t-table or using statistical software, resulting in, for example, a margin of error of 1.98 cm.

Calculate the confidence interval: Take the sample mean (50 cm) and add/subtract the margin of error (1.98 cm). This yields a confidence interval of 48.02 cm to 51.98 cm.

Interpretation: With 95% confidence, we can say that the average height of the population lies between 48.02 cm and 51.98 cm based on our sample.

The confidence interval provides a range of plausible values for the population parameter, taking into account the uncertainty associated with estimating it from a sample. The width of the interval is influenced by the sample size, variability, and chosen confidence level.




    
```

```python
Q6. Use Bayes' Theorem to calculate the probability of an event occurring given prior knowledge of the
event's probability and new evidence. Provide a sample problem and solution.



ANs:-
 
Certainly! Let's go through an example of using Bayes' theorem to calculate the probability of an event given prior knowledge and new evidence.

Example problem:
Suppose there is a rare disease that affects 1 in 10,000 people. There is a diagnostic test for this disease that has a 99% accuracy rate, meaning it correctly identifies the presence of the disease in 99% of cases where it is present (true positive rate) and correctly identifies the absence of the disease in 99% of cases where it is not present (true negative rate).

Now, imagine a person tests positive for the disease using this test. We want to calculate the probability that the person actually has the disease.

Solution:
To apply Bayes' theorem, we need the following information:

P(A): Prior probability of having the disease = 1/10,000 = 0.0001
P(B|A): Probability of testing positive given the person has the disease = 0.99 (true positive rate)
P(B|not A): Probability of testing positive given the person does not have the disease = 1 - 0.99 = 0.01 (false positive rate)
First, let's calculate the probability of testing positive (P(B)):

P(B) = P(B|A) * P(A) + P(B|not A) * P(not A)

P(B) = 0.99 * 0.0001 + 0.01 * (1 - 0.0001)

P(B) ≈ 0.010098

Now, we can apply Bayes' theorem to calculate the probability of having the disease given a positive test result (P(A|B)):

P(A|B) = (P(B|A) * P(A)) / P(B)

P(A|B) = (0.99 * 0.0001) / 0.010098

P(A|B) ≈ 0.0098

So, the probability that the person actually has the disease given a positive test result is approximately 0.0098 or 0.98%.

Despite testing positive, the low prevalence of the disease results in a relatively low probability of having the disease.

Bayes' theorem allows us to update our initial probability (prior knowledge) based on the observed test result (new evidence) to obtain the posterior probability. It demonstrates the importance of considering both the accuracy of the test and the prevalence of the disease when interpreting the results.




```

```python
Q10. In a study of the effectiveness of a new weight loss drug, a sample of 50 participants lost an average
of 6 pounds with a standard deviation of 2.5 pounds. Conduct a hypothesis test to determine if the drug is
significantly effective at a 95% confidence level using a t-test.

Ans:-
    
To conduct a hypothesis test to determine if the weight loss drug is significantly effective at a 95% confidence level using a t-test, we need to set up the null and alternative hypotheses, choose the appropriate significance level, calculate the test statistic, and compare it to the critical value or p-value.

Null hypothesis (H₀): The weight loss drug is not significantly effective (mean weight loss = 0).
Alternative hypothesis (H₁): The weight loss drug is significantly effective (mean weight loss ≠ 0).

Significance level (α): The chosen significance level is 0.05, corresponding to a 95% confidence level.

Now, let's calculate the test statistic using the t-test formula:

t = (sample mean - hypothesized mean) / (sample standard deviation / √sample size)

Given:
Sample size (n) = 50
Sample mean (x̄) = 6 pounds
Sample standard deviation (s) = 2.5 pounds

Hypothesized mean (μ₀) = 0 (since the null hypothesis assumes no weight loss)

t = (6 - 0) / (2.5 / √50)
t = 6 / (2.5 / 7.071)
t = 6 / 0.3542
t ≈ 16.94

Next, we need to compare the calculated t-value to the critical value or p-value. Since the sample size is large (n > 30), we can approximate the t-distribution with the standard normal distribution.

At a 95% confidence level, with a two-tailed test, the critical values are approximately ±1.96.

Since the calculated t-value (16.94) is much larger than the critical value (1.96), we can reject the null hypothesis.

Conclusion:
Based on the results of the t-test, there is sufficient evidence to conclude that the weight loss drug is significantly effective at a 95% confidence level. The average weight loss of 6 pounds in the sample is significantly different from zero, indicating that the drug has an effect on weight loss in the population.    
```

```python
Q11. In a survey of 500 people, 65% reported being satisfied with their current job. Calculate the 95%
confidence interval for the true proportion of people who are satisfied with their job.

Ans:=
    
 To calculate the 95% confidence interval for the true proportion of people who are satisfied with their job, we can use the formula:

Confidence Interval = Sample Proportion ± (Z * Standard Error)

where:

Sample Proportion is the observed proportion in the sample,
Z is the critical value for the desired confidence level (95% confidence level corresponds to a Z-value of approximately 1.96),
Standard Error is the standard deviation of the sampling distribution of the proportion, which can be calculated as √((p * (1-p)) / n), where p is the sample proportion and n is the sample size.
Given:
Sample Size (n) = 500
Sample Proportion (p) = 0.65 (65% satisfaction rate)
Z-value for 95% confidence level ≈ 1.96

Let's calculate the confidence interval:

Standard Error = √((p * (1-p)) / n)
Standard Error = √((0.65 * (1-0.65)) / 500)
Standard Error ≈ √(0.2275 / 500)
Standard Error ≈ √0.000455
Standard Error ≈ 0.0214

Confidence Interval = 0.65 ± (1.96 * 0.0214)
Confidence Interval = 0.65 ± 0.0419

The 95% confidence interval for the true proportion of people who are satisfied with their job is approximately (0.6081, 0.6919).

Interpretation:
We can say with 95% confidence that the true proportion of people who are satisfied with their job lies within the range of approximately 60.81% to 69.19% based on the given sample. This interval provides an estimated range for the true population proportion, taking into account the variability introduced by sampling.   
```
